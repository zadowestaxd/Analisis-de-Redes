```python
#Diámetro
print(nx.is_connected(G))

componentes = nx.connected_components(G)
componente_mayor = max(componentes,key=len)
componente_mayor

subgrafo = G.subgraph(componente_mayor)
diámetro = nx.diameter(subgrafo)
print("Diametro de la red de componentes mayormente conectados:", diámetro)
```

```python
#Centralidad
centralidad_intermediación = nx.betweenness_centrality(G)
centralidad_vector_propio = nx.eigenvector_centrality(G)
#Asignamos un atributo para cada centralidad calculada
nx.set_node_attributes(G,centralidad_intermediación,'intermediacion')
nx.set_node_attributes(G,centralidad_vector_propio,'v_propio')
#Creamos la lista ordenada de nodos más importantes por centralidad
import operator
centralidad_ordenado = sorted(centralidad_intermediación.items(), key=operator.itemgetter(1), reverse=True)
#Mostramos la lista de los 50 nodos más importantes por intermediación
print("Los 20 nodos más importantes según su centralidad por intermediación:")
for i in centralidad_ordenado[:50]:
    print(i)
```
