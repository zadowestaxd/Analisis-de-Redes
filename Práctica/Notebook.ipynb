```python
import networkx as nx
```

```python
#Importamos la red en la variable G añadiendo el atributo timestamp a las aristas.
#Este atributo indica el tiempo que han anidado juntos los tejedores representados por los vértices de la arista
G = nx.read_edgelist('aves-weaver-social.edges', nodetype=int, data=(('timestamp',float),))
```

```python
import matplotlib.pyplot as plt
```

```python
density = nx.density(G)
print("Densidad de la red:", density)
#La densidad de la red es bastante baja lo que nos permite saber que no hay una concentración
#alta de enlaces entre los nodos de la red comparado con el número máximo de enlaces que podría haber
#Podemos suponer que los sujetos al fin y al cabo no
#tienen tanta relación entre todos los miembros de la especie pero se relacionan en grupos. En 
#ese caso tendría sentido ya que los vínculos en la red estan distribuidos entre distintos 
#subgrupos sin concentrarse en un único punto.
```

```python
#Centralidad
centralidad_intermediación = nx.betweenness_centrality(G)
centralidad_vector_propio = nx.eigenvector_centrality(G)
#Asignamos un atributo para cada centralidad calculada
nx.set_node_attributes(G,centralidad_intermediación,'intermediacion')
nx.set_node_attributes(G,centralidad_vector_propio,'v_propio')
#Creamos la lista ordenada de nodos más importantes por centralidad
import operator
intermediacion_ordenado = sorted(centralidad_intermediación.items(), key=operator.itemgetter(1), reverse=True)
vpropio_ordenado = sorted(centralidad_vector_propio.items(), key=operator.itemgetter(1), reverse=True)
```

```python
#Mostramos la lista de los 30 nodos más importantes por centralidad de intermediación
print("Los 30 nodos más importantes según su centralidad por intermediación:")
for i in intermediacion_ordenado[:30]:
    print(i)
```

```python
#Mostramos la lista de los 30 nodos más importantes por centralidad por vector propio
print("Los 30 nodos más importantes según su centralidad de vector propio:")
for i in vpropio_ordenado[:30]:
    print(i)
#Estos obtienen su valor no solo por la cantidad de aristas de un nodo si no también 
#por la relevancia en la red de los nodos a los que está conectados. 
#Esto significa que estos tejedores han anidado en cámaras altamente utilizadas por otros miembros
```

```python
#Ahora vamos a ver la red completa con clasificación por colores de los nodos y tamaño según centralidad de intermediación
#y grosor de las aristas según el timestamp
vecinos=[] 
for i in G:
    lista = list(G.neighbors(i)) #Lista de vecinos del nodo i
    vecinos.append(len(lista))   #Contamos los elementos de la lista de vecinos de i y añadimos el numero de vecinos a vecinos[]
color_map = []
for i in G:
        if vecinos[i-1] < 2: #nodos con un solo vecino
            color_map.append('black')
        elif  vecinos[i-1] < 5:#nodos con menos vecinos que la media
            color_map.append('red')
        elif  vecinos[i-1] < 6:#nodos con la media de vecinos
            color_map.append('blue')
        elif vecinos[i-1]  < 15:#nodos por encima de la media con menos de 15 vecinos
            color_map.append('purple')
        else :#Nodos con mas de 15 vecinos
            color_map.append('green')
#Tamaño de los nodos según su c
node_size = [20 + (10000*nx.get_node_attributes(G, 'intermediacion')[v]) for v in G]
edge_width = [0.3 + 0.05*G[u][v]['timestamp'] for u,v in G.edges()]




```

```python
plt.figure(figsize=(15,10))
nx.draw_networkx(G, node_color=color_map, node_size=node_size, 
                 alpha=0.5, with_labels=False, 
                 width=edge_width, edge_color='.4')
```
