```python
import networkx as nx
```

```python
G = nx.read_edgelist('aves-weaver-social.edges', nodetype=int, data=(('timestamp',float),))
```

```python
import matplotlib.pyplot as plt
```

```python
G.nodes()
```

```python
plt.figure(figsize=(10,7))

# El alpha cuanto más bajo más transparente y en edge_color al revés.
nx.draw_networkx(G, alpha=0.5, node_size = 150, with_labels=True, font_size= 8, width= 2, edge_color='0.5')

#plt.axis('off')
plt.tight_layout();
```

```python
density = nx.density(G)
print("Densidad de la red:", density)
#La densidad de la red es bastante baja lo que nos permite saber que no hay una concentración
alta de enlaces entre los nodos de la red. Podemos suponer que los sujetos al fin y al cabo no
tienen tanta relación entre todos los miembros de la especie pero se relacionan en grupos. En 
ese caso tendría sentido ya que los vínculos en la red estan distribuidos entre distintos 
subgrupos sin concentrarse en un mismo punto.
```
