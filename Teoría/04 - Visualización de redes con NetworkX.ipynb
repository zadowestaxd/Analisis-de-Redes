# Visualización de redes con NetworkX

```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt
```

Vamos a importar el archivo major_us_cities que tiene un grafo serializado en el formato pickle de Python (pickle es un formato de serializar objetos de Python, es decir, de transformarlos en un stream de bytes). 

El archivo tiene 51 ciudades de Estados Unidos, que serán los nodos y que tienen dos atributos: location y population. 

Las aristas tienen el valor  del coste de desplazamiento entre ambas ciudades.

```python
# Leer el archivo directamente en un grafo
G = nx.read_gpickle('major_us_cities')

```

```python
# Vemos la información que tienen los nodos
G.nodes(data=True)
```

```python
# Vemos la información que tienen las aristas
G.edges(data=True)
```

## Poniendo una posición específica para los nodos

Probamos primero la disposición de la red normal

```python
# draw the graph using the default spring layout
plt.figure(figsize=(10,9))
nx.draw_networkx(G)
```

Usamos la posición que viene en el atributo location y que coincide con la situación de las ciudades en el mapa del país

```python

plt.figure(figsize=(10,7))

pos = nx.get_node_attributes(G, 'location')
nx.draw_networkx(G, pos)
```

## Modificando el aspecto del gráfico

A continuación vemos como modificar la transparencia del nodo (`alpha`) y la intensidad del color de los ejes (`edge_color`), así como eliminar los nombres de los nodos

```python

plt.figure(figsize=(10,7))

# El alpha cuanto más bajo más transparente y en edge_color al revés.
nx.draw_networkx(G, pos, alpha=0.5, with_labels=False, edge_color='0.3')

#plt.axis('off')
plt.tight_layout();
```

A continuación asociaremos lo siguiente:
- El grado del nodo con la intensidad de su color (`edge_color`)
- La población de la ciudad con el tamaño del nodo (`node_size`)
- El grosor de la arista con el peso de la misma (`edge_width`) 

Dejaremos la transparencia del nodo y la intensidad de los ejes como antes y usaremos el mapa de color viridis ([puedes ver los disponibles aquí](https://matplotlib.org/stable/gallery/color/colormap_reference.html))

```python
plt.figure(figsize=(10,7))

node_color = [G.degree(v) for v in G]
node_size = [0.0005*nx.get_node_attributes(G, 'population')[v] for v in G]
edge_width = [0.0015*G[u][v]['weight'] for u,v in G.edges()]

nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=0.7, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.viridis)

#plt.axis('off')
plt.tight_layout();
```

Repetimos la configuración anterior de la red, pero modificando el grosor y el color de las aristas, y añadiendo la etiqueta en algunos nodos

```python
plt.figure(figsize=(10,7))

# Repetimos la configuración de la red anterior
node_color = [G.degree(v) for v in G]
node_size = [0.0005*nx.get_node_attributes(G, 'population')[v] for v in G]
edge_width = [0.0015*G[u][v]['weight'] for u,v in G.edges()]

nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=0.7, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.viridis)



# Obtenemos aquellas aristas que tienen un coste superior a 770
# Sacamos una lista de booleanas con aquellas que cumplen esa condición
greater_than_770 = [x for x in G.edges(data=True) if x[2]['weight']>770] # El tercer elemento del eje es el diccionario de atributos

# Le cambiamos el grosor y color especial a las aristas que cumplen esa condición
nx.draw_networkx_edges(G, pos, edgelist=greater_than_770, edge_color='r', alpha=0.4, width=6)

# Para mostrar las etiquetas, mostramos un diccionario que mapee el ID del nodo con su etiqueta 
# sólo mostraremos LA y NYC)
nx.draw_networkx_labels(G, pos, labels={'Los Angeles, CA': 'LA', 'New York, NY': 'NYC'}, font_size=18, font_color='w')

#plt.axis('off')
plt.tight_layout();
```
