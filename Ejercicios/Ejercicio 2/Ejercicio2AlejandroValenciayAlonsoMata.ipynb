```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import csv

G = nx.read_edgelist("edgelist.txt",  delimiter=",",nodetype = str,data=[('intensidad', int), ('forma', int)])
#edgeslist = pd.read_csv('edgelist.txt')
#for i,j in edgeslist.items():
#       for jj in j:
 #            G.add_edge(i, jj, data= (("intensidad", int),("forma", int)))
#edges = nx.read_edgelist(edgeslist, data= (("intensidad", int),("forma", int)))
#G.add_edges_from(edges.edges())
G.edges(data=True)

csvFile = pd.read_csv('9_11_attrib.txt')
G.add_nodes_from(csvFile)
## for i in csvFile:
## my_graph.add_node(i)
G.edges(data=True)
G.nodes(data = True)

plt.figure(figsize=(32,28))
node_color = [G.degree(v) for v in G]
edge_width = [0.6*G[u][v]['intensidad'] for u,v in G.edges()]
plt.tight_layout();
nx.draw_networkx(G,
                 node_color=node_color, alpha=0.7, with_labels=True, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.viridis)
```
