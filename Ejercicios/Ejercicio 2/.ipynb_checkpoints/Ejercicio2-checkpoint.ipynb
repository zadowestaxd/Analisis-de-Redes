```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas

G = nx.Graph()
edges = nx.read_edgelist('edgelist.txt', data= ((("intensidad", int))))
G.add_edges_from(edges.edges())
nx.draw(G, with_labels=False, font_weight='bold')
G.edges()
```
